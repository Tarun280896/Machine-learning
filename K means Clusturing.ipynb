```python
import pandas as pd
import numpy as np 
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
x = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
for i in range(1,21):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    print(kmeans.inertia_)
```

```python
centroid =[]
for i in range(1,15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    centroid.append(kmeans.inertia_)
```

```python
from matplotlib import pyplot as ptl
```

```python
plt.plot(np.arange(1,15),centroid,marker='o')
```

```python
kmeans = KMeans(n_clusters=5,init= 'k-means++',random_state=42)
```

```python
pred =kmeans.fit_predict(x)
```

```python
len(pred)
```

```python
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],cmap='black')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```
